# Creating a DF

```python
# Make my own DF
Name = ['Tom', 'Jane','Alice']
Age = [20,21,19]
Major = ["Business","Data Analytics","Business"]

toDF = {"Name": Name, "Age": Age, "Major": Major}

DF = pd.DataFrame(toDF)
print(len(DF))
DF.head()
```

```python
# With Lists
toDF2 = [["Tom", 20, "Business"], ["Jane", 21, "Data Analytics"], ["Alice", 19, "Business"]]
DF2 = pd.DataFrame(toDF2, columns = ["Name", "Age", "Major"])
DF2.head()
```
